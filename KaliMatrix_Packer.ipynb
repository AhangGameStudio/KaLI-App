# Kali Matrix APK 打包脚本 (Google Colab 版)

由于本地 Windows 环境打包 Buildozer 极其复杂且容易失败，使用 Google Colab 是最快、最稳定的解决方案。

## 使用方法
1. 上传文件：
   点击左侧文件夹图标，将你电脑上的以下文件上传到 Colab 的当前目录：
   - `mobile_app.py`
   - `data.py`
   - `buildozer.spec`
   - `DroidSansFallback.ttf` (可选，如果需要字体支持)

2. 依次运行下面的代码块。
3. 等待打包完成，最后下载生成的 APK 文件。

```python
# 1. 安装必要的系统依赖和 Buildozer
!sudo apt-get update
!sudo apt-get install -y python3-pip build-essential git python3 python3-dev ffmpeg libsdl2-dev libsdl2-image-dev libsdl2-mixer-dev libsdl2-ttf-dev libportmidi-dev libswscale-dev libavformat-dev libavcodec-dev zlib1g-dev
!sudo apt-get install -y libgstreamer1.0
!sudo apt-get install -y build-essential libsqlite3-dev sqlite3 bzip2 libbz2-dev zlib1g-dev libssl-dev openssl libgdbm-dev libgdbm-compat-dev liblzma-dev libreadline-dev libncursesw5-dev libffi-dev uuid-dev libffi6
!sudo apt-get install -y libffi-dev

!pip install --upgrade buildozer cython virtualenv
```

```python
# 2. 开始打包 (首次运行需要下载大量 Android SDK/NDK 文件，请耐心等待)
!buildozer android debug
```

```python
# 3. 打包完成后，列出并下载 APK 文件
import os
from google.colab import files

bin_dir = "bin"
if os.path.exists(bin_dir):
    files_in_bin = os.listdir(bin_dir)
    for f in files_in_bin:
        if f.endswith(".apk"):
            print(f"Found APK: {f}")
            files.download(os.path.join(bin_dir, f))
else:
    print("没有找到 bin 目录，打包可能失败了，请检查上一步的输出日志。")
```
